> A **análise de correlação** é uma técnica estatística usada para medir e **descrever a força e a direção de uma relação entre duas variáveis**. Ela ajuda a entender se, e como, duas variáveis estão relacionadas, o que pode ser muito útil em várias áreas de análise de dados.

### **Principais Conceitos da Análise de Correlação**

- **Correlação Positiva**: Quando uma variável aumenta, a outra também tende a aumentar.
- **Correlação Negativa**: Quando uma variável aumenta, a outra tende a diminuir.
- **Correlação Nula**: Não há relação clara entre as duas variáveis.
- **Coeficiente de Correlação**: Um número entre -1 e 1 que quantifica a força e a direção da relação entre as variáveis. Os principais coeficientes são:
  - **Coeficiente de Correlação de Pearson**: Mede a relação linear entre duas variáveis.
  - **Coeficiente de Correlação de Spearman**: Mede a relação monotônica entre duas variáveis, útil para dados ordinais.
  - **Coeficiente de Correlação de Kendall**: Outra medida para dados ordinais, mais robusta para pequenas amostras.


# **Técnicas de analise de correlação**

## **1. Gráfico de Dispersão**

Os gráficos de dispersão ajudam a visualizar a relação entre duas variáveis.


```python
# @title exemplo
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Gerando dados fictícios
np.random.seed(42)
x = np.random.normal(50, 10, 100)
y = 2 * x + np.random.normal(0, 5, 100)  # Correlação positiva

# Criando um DataFrame
df = pd.DataFrame({'x': x, 'y': y})

# Plotando o gráfico de dispersão
plt.figure(figsize=(4, 3))
sns.scatterplot(x='x', y='y', data=df)
plt.title('Gráfico de Dispersão')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

## **2. Coeficiente de Correlação de Pearson**

O que é: Mede a força e a direção da relação linear entre duas variáveis contínuas.

**Método:** Calcula a relação linear entre as variáveis usando a fórmula do coeficiente de correlação de Pearson (r), que varia de -1 a 1.

- **+1 indica uma correlação positiva perfeita** (quando uma variável aumenta, a outra também aumenta de maneira linear).
- **-1 indica uma correlação negativa perfeita** (quando uma variável aumenta, a outra diminui de maneira linear).
- **0 indica nenhuma correlação linear.**
Supondo que: As variáveis sejam contínuas e a relação entre elas seja linear. A relação deve seguir uma distribuição normal.

```python
# @title exemplo
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Gerando dados fictícios
np.random.seed(42)
x = np.random.normal(50, 10, 100)
y = 2 * x + np.random.normal(0, 5, 100)  # Correlação positiva

# Criando um DataFrame
df = pd.DataFrame({'x': x, 'y': y})

# Calculando a correlação de Pearson
correlation = df.corr(method='pearson')
print("Correlação de Pearson:\n", correlation)

# Plotando a matriz de correlação
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Matriz de Correlação de Pearson')
plt.show()
```


## **3. Coeficiente de Correlação de Spearman**

O coeficiente de correlação de Spearman mede a relação monotônica entre duas variáveis.

>> **Método:** Usa os rankings das variáveis em vez dos valores brutos para calcular a correlação. O coeficiente de correlação de Spearman (ρ ou rs) varia de -1 a 1.
- **+1** indica uma correlação monótona positiva perfeita (quando uma variável aumenta, a outra também aumenta de maneira monotônica).
- **-1** indica uma correlação monótona negativa perfeita (quando uma variável aumenta, a outra diminui de maneira monotônica).
- **0** indica nenhuma correlação monotônica.


```python
# @title exemplo
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Gerando dados fictícios
np.random.seed(42)
x = np.random.normal(50, 10, 100)
y = 2 * np.log(x + 1) + np.random.normal(0, 5, 100)  # Correlação monotônica

# Criando um DataFrame
df = pd.DataFrame({'x': x, 'y': y})

# Calculando a correlação de Spearman
correlation_spearman = df.corr(method='spearman')
print("Correlação de Spearman:\n", correlation_spearman)

# Plotando a matriz de correlação
sns.heatmap(correlation_spearman, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Matriz de Correlação de Spearman')
plt.show()
```

## **4. Coeficiente de Correlação de Kendall**

O coeficiente de correlação de Kendall é usado para medir a relação entre duas variáveis ordinais.


```python
# @title exemplo
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Gerando dados fictícios
np.random.seed(42)
x = np.random.normal(50, 10, 100)
y = np.floor(2 * x + np.random.normal(0, 5, 100))  # Correlação ordinal

# Criando um DataFrame
df = pd.DataFrame({'x': x, 'y': y})

# Calculando a correlação de Kendall
correlation_kendall = df.corr(method='kendall')
print("Correlação de Kendall:\n", correlation_kendall)

# Plotando a matriz de correlação
sns.heatmap(correlation_kendall, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Matriz de Correlação de Kendall')
plt.show()
```
